```python
from sklearn.datasets import make_blobs
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import matplotlib.colors
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=10000, centers=5, n_features=2,random_state=10,)
```

```python
X
```

```python
y
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

SVM

```python
 temp_x,temp_y=np.hsplit(X_train,2)
```

```python
model = svm.SVC()
model.fit(X_train,y_train)
```

```python
colors = ["red", "blue","green","brown","skyblue"]
colormap = matplotlib.colors.ListedColormap(colors)
plt.scatter(temp_x, temp_y, c=y_train, cmap=colormap)
```

```python
prediction =  model.predict(X_test)
acc = accuracy_score(y_test,prediction)
# print ("predictions: ",prediction)
# print ("actual: " ,y_test)
print ("accuracy: ",acc*100 )
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(X_train,y_train)
KNN_pred = knn.predict(X_test)
acc_KNN = accuracy_score(y_test,KNN_pred)
print ("predictions: ",KNN_pred)
print ("actual: " ,y_test)
print ("accuracy: ",acc_KNN*100 )
```

Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python

dtree = DecisionTreeClassifier(max_depth=10,random_state=101,max_features=None,min_samples_leaf=15  )
dtree.fit(X_train,y_train)
dtree_pred = dtree.predict(X_test)
acc_dtree= accuracy_score(y_test,dtree_pred)
print ("predictions: ",dtree_pred)
print ("actual: " ,y_test)
print ("accuracy: ",acc_dtree*100)
```

Logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=500)
lr.fit(X_train,y_train)
lr_pred = lr.predict(X_test)
lr_acc = accuracy_score(y_test,lr_pred)
print ("predictions: ",lr_pred)
print ("actual: " ,y_test)
print ("accuracy: ",lr_acc*100)
```

Actual figure

```python
colors = ["red", "blue","green","brown","skyblue"]
colormap = matplotlib.colors.ListedColormap(colors)
plt.scatter(temp_x, temp_y, c=y_train, cmap=colormap)
```

```python
from sklearn.cluster import DBSCAN
dbs = DBSCAN(eps=1,min_samples=15)
y_pred = dbs.fit_predict(X_train)
plt.scatter(temp_x, temp_y, c=y_pred)
len(set(y_pred))
```

```python
from sklearn.cluster import KMeans
KM = KMeans(n_clusters=5)
y_pred = KM.fit_predict(X_train)
plt.scatter(temp_x, temp_y, c=y_pred)
len(set(y_pred))
```
